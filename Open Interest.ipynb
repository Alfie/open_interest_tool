```python
import robin_stocks as rs
import pandas as pd

#rs.options.get_option_historicals(***) this may be useful for quantdom

login = rs.login('alfredokang@gmail.com','Coolio!123')
ticker = 'slv'

rs.options.get_chains(ticker)

exp_dates = []
i=0
for e in rs.options.get_chains(ticker)['expiration_dates']:
    exp_dates.append(e)
    i+=1

strikes = []

for e in rs.options.find_options_by_expiration(ticker, exp_dates[0],optionType='call'):
    strikes.append(e['strike_price'])

def cast_list(test_list, data_type): 
    return list(map(data_type, test_list))

strikes = cast_list(strikes, float)
strikes.sort()
strikes = cast_list(strikes, str)

print(strikes)    
```

```python
for e in exp_dates:
    print(e)
    for s in strikes:
        options_at_strike = pd.DataFrame(rs.options.get_option_market_data(ticker,e,s,'put'))
        print(s)
        print(options_at_strike[['volume','open_interest']])
```
